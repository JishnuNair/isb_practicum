# Initial Analysis of Prices

```python
import pandas as pd
import os
import glob
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.graph_objs as go
#from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

%matplotlib inline
#init_notebook_mode()
```

```python
class cd:
    """Context manager for changing the current working directory"""
    def __init__(self, newPath):
        self.newPath = os.path.expanduser(newPath)

    def __enter__(self):
        self.savedPath = os.getcwd()
        os.chdir(self.newPath)

    def __exit__(self, etype, value, traceback):
        os.chdir(self.savedPath)

data_path='/home/jishnu/Documents/ISB/Term3/practicum/workspace/data_collection/data/daily_data'
with cd(data_path):
    listings_file = max(glob.iglob('listings_*'), key=os.path.getctime)
    ticker_file = max(glob.iglob('ticker_data_*'), key=os.path.getctime)
    global_file = max(glob.iglob('global_data_*'), key=os.path.getctime)

listings_file = os.path.join(data_path,listings_file)
ticker_file = os.path.join(data_path,ticker_file)
global_file = os.path.join(data_path,global_file)
```

```python
listings = pd.read_csv(listings_file)
ticker_data = pd.read_csv(ticker_file)
global_data = pd.read_csv(global_file)
```

## Plotting price variations for top 5 currencies for past year

```python
hist_dir = '/home/jishnu/Documents/ISB/Term3/practicum/workspace/data_collection/data/historical_data/prices'
bitcoin_hist = pd.read_csv('{0}/bitcoin.csv'.format(hist_dir))
ethereum_hist = pd.read_csv('{0}/ethereum.csv'.format(hist_dir))
ripple_hist = pd.read_csv('{0}/ripple.csv'.format(hist_dir))
bitcoin_cash_hist = pd.read_csv('{0}/bitcoin_cash.csv'.format(hist_dir))
eos_hist = pd.read_csv('{0}/eos.csv'.format(hist_dir))
```

```python
bitcoin_hist['Date'] = pd.to_datetime(bitcoin_hist['Date'])
bitcoin_hist['Date'] = bitcoin_hist['Date'].apply(lambda x:x.to_pydatetime())
ethereum_hist['Date'] = pd.to_datetime(ethereum_hist['Date'])
ethereum_hist['Date'] = ethereum_hist['Date'].apply(lambda x:x.to_pydatetime())
ripple_hist['Date'] = pd.to_datetime(ripple_hist['Date'])
ripple_hist['Date'] = ripple_hist['Date'].apply(lambda x:x.to_pydatetime())
bitcoin_cash_hist['Date'] = pd.to_datetime(bitcoin_cash_hist['Date'])
bitcoin_cash_hist['Date'] = bitcoin_cash_hist['Date'].apply(lambda x:x.to_pydatetime())
eos_hist['Date'] = pd.to_datetime(eos_hist['Date'])
eos_hist['Date'] = eos_hist['Date'].apply(lambda x:x.to_pydatetime())
```

## Plotting trend variations for top 5 currencies for past year

```python
hist_dir='/home/jishnu/Documents/ISB/Term3/practicum/workspace/data_collection/data/historical_data/sentiments'
gtrends_bitcoin = pd.read_csv('{0}/bitcoin_sentiment.csv'.format(hist_dir))
gtrends_ethereum = pd.read_csv('{0}/ethereum_sentiment.csv'.format(hist_dir))
gtrends_ripple = pd.read_csv('{0}/ripple_sentiment.csv'.format(hist_dir))
gtrends_bitcoin_cash = pd.read_csv('{0}/bitcoin_cash_sentiment.csv'.format(hist_dir))
gtrends_eos = pd.read_csv('{0}/EOS_sentiment.csv'.format(hist_dir))
```

```python
gtrends_bitcoin.reset_index(level=0, inplace=True)
gtrends_bitcoin.columns = ['date','interest']
gtrends_bitcoin = gtrends_bitcoin.query('date != "Week"')
gtrends_bitcoin['date'] = pd.to_datetime(gtrends_bitcoin['date'])

gtrends_ethereum.reset_index(level=0, inplace=True)
gtrends_ethereum.columns = ['date','interest']
gtrends_ethereum = gtrends_ethereum.query('date != "Week"')
gtrends_ethereum['date'] = pd.to_datetime(gtrends_ethereum['date'])

gtrends_ripple.reset_index(level=0, inplace=True)
gtrends_ripple.columns = ['date','interest']
gtrends_ripple = gtrends_ripple.query('date != "Week"')
gtrends_ripple['date'] = pd.to_datetime(gtrends_ripple['date'])

gtrends_bitcoin_cash.reset_index(level=0, inplace=True)
gtrends_bitcoin_cash.columns = ['date','interest']
gtrends_bitcoin_cash = gtrends_bitcoin_cash.query('date != "Week"')
gtrends_bitcoin_cash['date'] = pd.to_datetime(gtrends_bitcoin_cash['date'])

gtrends_eos.reset_index(level=0, inplace=True)
gtrends_eos.columns = ['date','interest']
gtrends_eos = gtrends_eos.query('date != "Week"')
gtrends_eos['date'] = pd.to_datetime(gtrends_eos['date'])
```

```python
gtrends_bitcoin_data = [go.Scatter(x=gtrends_bitcoin['date'],y=gtrends_bitcoin['interest'])]
layout = go.Layout(
    title='Past Year Online Interest Variation for Bitcoin',
    xaxis=dict(title='Date',titlefont=dict(family='Courier New, monospace',size=18,color='#7f7f7f')),
    yaxis=dict(title='Web Interest',titlefont=dict( family='Courier New, monospace',size=18,color='#7f7f7f'))
)
fig = go.Figure(data=gtrends_bitcoin_data, layout=layout)

plotly.offline.iplot(fig)
```

```python
gtrends_ethereum_data = [go.Scatter(x=gtrends_ethereum['date'],y=gtrends_ethereum['interest'])]
layout = go.Layout(
    title='Past Year Online Interest Variation for Ethereum',
    xaxis=dict(title='Date',titlefont=dict(family='Courier New, monospace',size=18,color='#7f7f7f')),
    yaxis=dict(title='Web Interest',titlefont=dict( family='Courier New, monospace',size=18,color='#7f7f7f'))
)
fig = go.Figure(data=gtrends_ethereum_data, layout=layout)

plotly.offline.iplot(fig)
```

```python
gtrends_bitcoin_data = go.Scatter(x=gtrends_bitcoin['date'],y=gtrends_bitcoin['interest'],name='Bitcoin')
gtrends_ethereum_data = go.Scatter(x=gtrends_ethereum['date'],y=gtrends_ethereum['interest'],name='Ethereum')
gtrends_ripple_data = go.Scatter(x=gtrends_ripple['date'],y=gtrends_ripple['interest'],name='Ripple')
gtrends_bitcoin_cash_data = go.Scatter(x=gtrends_bitcoin_cash['date'],y=gtrends_bitcoin_cash['interest'],name='Bitcoin Cash')
gtrends_eos_data = go.Scatter(x=gtrends_eos['date'],y=gtrends_eos['interest'],name='EOS')

layout = go.Layout(
    title='Past Year Online Interest Variation for Top 5 CryptoCurrencies',
    xaxis=dict(title='Date',titlefont=dict(family='Courier New, monospace',size=18,color='#7f7f7f')),
    yaxis=dict(title='log(Web Interest)',titlefont=dict( family='Courier New, monospace',size=18,color='#7f7f7f'),type='log')
)
data=[gtrends_bitcoin_data,gtrends_ethereum_data,gtrends_ripple_data,gtrends_bitcoin_cash_data,gtrends_eos_data]
fig = go.Figure(data=data, layout=layout)

plotly.offline.iplot(fig)
```

On analyzing the trend of price(Pulled from coinmarket site) and online interest(using Google Trends data) for the top 5 cryptocurrencies, it is seen that there was a definite spike in prices and online interest for all currencies in the timeframe from **November, 2017 to January, 2018**.

It needs to be analyzed why the spike occurred. Was it all due to hype? Or was it due to "Pump and Dump" strategies?
