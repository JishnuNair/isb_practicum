# Twitter Sentiment Analysis

```python
import os
import glob
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.sentiment.util import *
from nltk import tokenize

from sklearn.feature_extraction.text import TfidfVectorizer

%matplotlib inline
```

```python
class cd:
    """Context manager for changing the current working directory"""
    def __init__(self, newPath):
        self.newPath = os.path.expanduser(newPath)

    def __enter__(self):
        self.savedPath = os.getcwd()
        os.chdir(self.newPath)

    def __exit__(self, etype, value, traceback):
        os.chdir(self.savedPath)

data_path='/home/jishnu/Documents/ISB/Term3/practicum/workspace/data_collection/data/daily_data'
with cd(data_path):
    listings_file = max(glob.iglob('listings_*'), key=os.path.getctime)
    ticker_file = max(glob.iglob('ticker_data_*'), key=os.path.getctime)
    global_file = max(glob.iglob('global_data_*'), key=os.path.getctime)
    tweets_file = max(glob.iglob('tweets_*'), key=os.path.getctime)

listings_file = os.path.join(data_path,listings_file)
ticker_file = os.path.join(data_path,ticker_file)
global_file = os.path.join(data_path,global_file)
tweets_file = os.path.join(data_path,tweets_file)
```

```python
listings = pd.read_csv(listings_file)
ticker_data = pd.read_csv(ticker_file)
global_data = pd.read_csv(global_file)
tweets = pd.read_table(tweets_file,sep='###!###',
                       header=None,names=['tweet_no','user','date','location','tweet'],
                       parse_dates=['date'],engine='python')
```

```python
tweets['tweet'].head(10)
```

## Preprocessing tweets

```python
#Preprocessing del RT @blablabla:
tweets['tweetos'] = '' 
tweets['tweet_clean'] = ''

#add tweetos first part
for i in range(len(tweets['tweet'])):
    try:
        tweets.iloc[i,5] = tweets['tweet'].str.split(' ')[i][1]
    except AttributeError:    
        tweets.iloc[i,5] = 'other'

#Preprocessing tweetos. select tweetos contains 'b@'
for i in range(len(tweets['tweet'])):
    if tweets['tweetos'].str.contains('@')[i]  == False:
        tweets.iloc[i,5] = 'other'
        
# remove URLs, RTs, and twitter handles
for i in range(len(tweets['tweet'])):
    tweets.iloc[i,6] = " ".join([word for word in tweets['tweet'][i].split()
                                if 'http' not in word and '@' not in word and '<' not in word])
```

```python
tweets.head(5)
```

```python
tweets['tweet_clean'] = tweets['tweet_clean'].apply(lambda x: re.sub('[!@#$:).;,?&]', '', x.lower()))
tweets['tweet_clean'] = tweets['tweet_clean'].apply(lambda x: re.sub('  ', ' ', x))
```

## Plotting word clouds

```python
def wordcloud(tweets,col):
    stopwords = set(STOPWORDS)
    wordcloud = WordCloud(background_color="white",stopwords=stopwords,random_state = 2016).generate(" ".join([i for i in tweets[tweets[col].notnull()][col]]))
    plt.figure( figsize=(16,9), facecolor='k')
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.title("Word Cloud")
wordcloud(tweets,'tweet_clean')  
```

```python
wordcloud(tweets,'location')
```

## Running Sentiment Analysis

```python
tweets['text_lem'] = [''.join([WordNetLemmatizer().lemmatize(re.sub('[^A-Za-z]', ' ', line)) for line in lists]).strip() for lists in tweets['tweet_clean']]       
vectorizer = TfidfVectorizer(max_df=0.5,max_features=10000,min_df=10,stop_words='english',use_idf=True)
X = vectorizer.fit_transform(tweets['text_lem'].str.upper())
sid = SentimentIntensityAnalyzer()
tweets['sentiment_compound_polarity']=tweets.text_lem.apply(lambda x:sid.polarity_scores(x)['compound'])
tweets['sentiment_neutral']=tweets.text_lem.apply(lambda x:sid.polarity_scores(x)['neu'])
tweets['sentiment_negative']=tweets.text_lem.apply(lambda x:sid.polarity_scores(x)['neg'])
tweets['sentiment_pos']=tweets.text_lem.apply(lambda x:sid.polarity_scores(x)['pos'])
tweets['sentiment_type']=''
tweets.loc[tweets.sentiment_compound_polarity>0,'sentiment_type']='POSITIVE'
tweets.loc[tweets.sentiment_compound_polarity==0,'sentiment_type']='NEUTRAL'
tweets.loc[tweets.sentiment_compound_polarity<0,'sentiment_type']='NEGATIVE'
```

```python
tweets.head()
```

```python
tweets_sentiment = tweets.groupby(['sentiment_type'])['sentiment_neutral'].count()
tweets_sentiment.rename("",inplace=True)
explode = (1, 0, 0)
plt.subplot(221)
tweets_sentiment.transpose().plot(kind='barh',figsize=(20, 20))
plt.title('Sentiment Analysis 1', bbox={'facecolor':'0.8', 'pad':0})
plt.subplot(222)
tweets_sentiment.plot(kind='pie',figsize=(20, 20),autopct='%1.1f%%',shadow=True,explode=explode)
plt.legend(bbox_to_anchor=(1, 1), loc=3, borderaxespad=0.)
plt.title('Sentiment Analysis 2', bbox={'facecolor':'0.8', 'pad':0})
plt.show()
```
